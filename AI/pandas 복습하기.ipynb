pandas의 read_csv를 사용하여 csv(' ","로 나눠진 데이터셋 파일 ')를 읽어들일 수 있다. 반환되는 데이터 타입은 pd.DataFrame 타입이다.

첫번째 파라미터로 csv파일의 위치(str)를 받으며, sep(데이터를 나누는 문자)를 설정함으로서 csv 외의 파일도 불러올 수 있을 것으로 보인다. encoding 파라미터를 수정하여 들어오는 데이터의 인코딩 형식으로 데이터를 읽어올 수 있다.

```python
import pandas as pd

data = pd.read_csv('../input/korean-real-estate-transaction-data/Apart Deal.csv', encoding='utf-8', na_values=' ')
```

### loc은 DataFrame의 column이름이나 index이름으로 데이터를 불러온다.

```python
data.loc[2:5, '거래일': '지번']
```

조건형으로 원하는 데이터만을 불러올 수도 있다.

```python
data.loc[(data['건축년도'].isin([2017, 2019])) & (data['거래금액'] >= 400000)]
```

```python
def remean_points(row):
    row['건축년도'] = row['건축년도'] - int(data['건축년도'].mean())
    return row
```

```python
data.iloc[:100].apply(remean_points, axis='columns')
```
